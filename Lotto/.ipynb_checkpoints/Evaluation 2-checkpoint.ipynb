# Cross Validation and More Errors

Today we will finish up with various model evaluation metrics. We will introduce cross validation, one of the most important ways for finding the parameters for models. We will look at the decomposition of model errors into standard components. FInally we will see some more measures beyond accuracy scores for classification problems.

As always let's load some packages we need:

```python
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt

import random
np.random.seed(123)
random.seed(123)
```

```python
def PlotBoundaries(model, X) :
    '''
    Helper function that plots the decision boundaries of a model and data (X,Y)
    '''
    
    x_min, x_max = X[:, 0].min(), X[:, 0].max()
    y_min, y_max = X[:, 1].min(),X[:, 1].max()
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.4)
    plt.show()
```

```python
def compare_models(model_1,model_2):
    x_min, x_max = -1, 1
    y_min, y_max = -1, 1

    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

    Z  = np.c_[xx.ravel(), yy.ravel()]
    return np.sum(model_1.predict(Z) == model_2.predict(Z)) / len(Z)
```

## Cross Validation

Recall, we divided our data into 3 groups train, validation and test. Each had their own use:

1. Train: We fit the model to this data Using this we adjust the underlying parameters in a model, like the weights in logsitic regression. As we've seen we almost always overfit, even slightly, to the training data.
2. Validation: Used to tune a model's hyper parameters, such as the regularization penalty in logistic regression. This serves as a somewhat unbiased estimate of the model's performance. Since the model is not fitted on this data it should not overfit when using the `fit` parameter to it. But since we are adjusting the hyper parametres on this set we still implicitly overfit to it.
3. Test: Used to provide a totally unbiased estimate of the model's performance. This is data the model has never seen, it should serve as a good predictor for the model's performance once deployed and making prediction on new data.

As we've seen many times our model can easily overfit, even slightly, to the data in the training and validation sets. Differences in how the data is divided between these two sets can impact what models we end up selecting. For example consider the following data, that we divide into two training and validation sets:

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
wine = load_wine()
X = wine.data
Y = wine.target

X_train_one, X_test_one, y_train_one, y_test_one  = train_test_split(X,Y, train_size = 0.8, random_state = 1)
X_train_two, X_test_two, y_train_two, y_test_two = train_test_split(X,Y, train_size = 0.8, random_state = 123)
```

Each pair of sets contains the exact same data, but how it is split between their respective training and validation sets is different. This can have a huge impact on model selection. Say we wanted to deploy a KNN model. We can fit the model to each  training set and evaluate it on the respective validation set, finding the best $K$ for each:

```python
from sklearn.neighbors import KNeighborsClassifier

def find_best_K(X_train,Y_train,X_validate,Y_validate) :
    
    k_range = [i for i in range(100)]
    validation_score = []
    for K in k_range :
        K += 1
        validation_score.append(KNeighborsClassifier(n_neighbors = K).fit(X_train,Y_train).score(X_validate,Y_validate))
    return validation_score.index(max(validation_score)) + 1

print("Best K for first split:", find_best_K(X_train_one, y_train_one, X_test_one, y_test_one))
print("Best K for second split:", find_best_K(X_train_two, y_train_two, X_test_two, y_test_two))
```

We got different $K$ values!

How we divide our data between the train and validation set can have a huge impact on model selection and hence model performance. How can we be sure the split we used was a good one? In general we can't. How do we get around this? Instead of trying out a particular split we can try every split. 

This is the idea behind *cross validation*. We try every divide of data between train and validation. We select the model which got the best performance on average across all validation sets. While this is a nice sounding idea it's totally impractical. If we have $n$ data points there are $2^n-1$ ways to divide the data between the two sets. At $n=1000$ this already exceeds the number of atoms in the known universe. 

### KFold Cross Validation

In practice instead of trying every split we try out a subset of the possible splits. We first partition the data into $K$ groups. We try every division between train and validation where the training set contains $K-1$ of the groups while the validation set contains the one remaining group. To evaluate our models we look at the average over all the validation sets. Since we're limmiting ourselves to validation sets consisting of only one group we only have $K$ potential splits. This is called KFold cross validation:

<img src="http://i.imgur.com/pMepZ3V.jpg",width=550,height=550>

KFold cross validation is implemented in scikit-learn using the `cross_val_score` method under `sklearn.model_selection`.

```python
from sklearn.model_selection import cross_val_score
```

Let's use the cancer data to experiment with KFold cross-validation. We'll also reserve some data for a test set:

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X = cancer.data
Y = cancer.target

X, X_test, y, y_test  = train_test_split(X,Y, train_size = 0.8, random_state = 1)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

We can see the cross validation scores of a particular model, let's say KNN (with the default settings):

```python
from sklearn.tree import DecisionTreeClassifier
KNN_model = KNeighborsClassifier()

#Let's do 10 folds.
scores = cross_val_score(KNN_model, X, y, cv = 10)
for i in range(len(scores)) :
    print("Fold %i: %3.3f" % (i ,scores[i]))
```

To actually use cross validation for model selection we need to evaluate each model we are considering on each of the folds, taking the one with the highest average. First let's take a noisy version of the cancer dataset:

```python
from sklearn.linear_model import LogisticRegression

from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X = cancer.data
Y = cancer.target

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

Now let's fit a logistic regression model. We'll cross validate over the regularization constant `C`. Recall, the higher `C` is the less regularized the model is.

```python
#Store the results
validation_score_list = []

sample_range = [10**i for i in np.arange(-7,7,0.25)]

#Do some cross validation
for i in sample_range :
    LR_model = LogisticRegression(C=i)
    validation_score = np.mean(cross_val_score(LR_model, X, Y, cv = 10))
    validation_score_list.append(validation_score)

plt.scatter(np.log10(sample_range), validation_score_list,label="Validation Score")
plt.legend();
plt.show();
```

Based on this we would take a regularization value somehwere near the middle range of values we considered.

### What K to choose?

The choice of K in K-Fold can have a huge impact on the choice of model, consider the following example:

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Store the results
validation_score_list = []

sample_range = np.arange(1,20)

#Do some cross validation

for i in sample_range :
    model = KNeighborsClassifier(n_neighbors=i)
    validation_score = np.mean(cross_val_score(model, X, Y, cv = 2))
    validation_score_list.append(validation_score)

plt.scatter(sample_range, validation_score_list,label="Validation Score")
plt.legend();
plt.show();

print("Best Model: ", validation_score_list.index(np.max(validation_score_list))+1)
```

```python
#Store the results
validation_score_list = []

sample_range = np.arange(1,20)

#Do some cross validation

num_yes = np.sum(np.where(y==1,1,0))
num_no = np.sum(np.where(y!=1,1,0))
max_folds = np.min([num_yes,num_no])

for i in sample_range :
    model = KNeighborsClassifier(n_neighbors=i)
    validation_score = np.mean(cross_val_score(model, X, Y, cv = max_folds))
    validation_score_list.append(validation_score)

plt.scatter(sample_range, validation_score_list,label="Validation Score")
plt.legend();
plt.show();

print("Best Model: ", validation_score_list.index(np.max(validation_score_list))+1)
```

We found different neighbor amounts again... The whole point of cross validation was to avoid variations in our model brough about by different training/validation splits. What went wrong? First let's look a bit closer at the data set:

```python
X.shape
```

There are around 600 data points, in the first experiment we ran 2-Fold cross validation. This means for each hyper parameter setting we're trying we train and test on sets of around 300 points each, and only do this twice. In the second experiment we ran around 200-Fold cross validation. This leads to 200 folds with only 3 data points each. For each setting of hyper parameters we trained with sets of size almost 600 and tested on sets of size 3. 

Both of these represent extremes of cross validation, one uses a few massive folds and one uses several small folds. Both are probably not optimal for getting good models. The first is barely any better than getting one random train/validation set split while the second has tiny and unrealistic test sets.

So what is the ideal number of folds? As with many things, really almost anything, in machine learning we don't have a perfect answer. A good rule of thumb is keeping $20\%$ of your data in each fold, or 5-Fold cross validation. But if we have a ton of data, say tens or hundreds of thousands of points we could probably get away with smaller folds. Remember at the end of the day we want each fold to be a good estimate for the general data distribution. On the flip side if we have very little data even splitting into two folds could lead to training sets which are too small, in this case we're probably best just doing a train/validation split at 70/30.

Another thing to consider is time. Cross validation is already an expensive procedure, compared to regular validation, if we need to grid search over a wide range of hyper parameter combinations we can end up spending a ton of time just choosing our models. Different amounts of folds will impact the model discovery time (this is an exercise for you). 

Despite these meta issues doing simple 2 or even 3 fold cross validation tends to be more sound than a normal train/validation split.

### Bias Variance Decomposition

Now we will look at where our model errors come from. We will decompose our error into two (actually three) terms. This section is a bit theoretical so we will take it slowly.

What are we doing when fitting models? We're trying to learn a relation between the dependent variable $Y$ and independent variable $X$. This relation is some arbitrary function $f$ (and some natural error $\epsilon$):

$$ Y = f(x) + \epsilon $$

When doing any statistical modeling we are estimating $f(x)$ using $\hat{f}(x)$, this $\hat{f}$ can be any arbitrary model, this includes KNN or SVMs for classification problems or linear regression for regression problems. The error in our model is the difference between the relation $f(x)$ and our model $\hat{f}(x)$. 

A good model should be close to the underlying relation while a bad model would be far away. When fitting our models we can't find the distance between our model and the true relation, if we knew the true relation we wouldn't need to bother modeling it. Instead we use some training and validation data to fit and evaluate our models with. The choice of model and choice of data can have a huge impact on our ability to estimate $f(x)$, a model which is too weak will fail to model the underlying relation one that is too strong will pick up on patterns only present in subsets of the data it is exposed too.

Consider the following example, each blue dot represents a realization of a model, fitted with different data, and the red dot represents the ideal relation:

```python
import matplotlib.gridspec as gridspec
np.random.seed(12345)

x = np.random.randn(10)
y = np.random.randn(10)

gridspec.GridSpec(2,2)

plt.subplot2grid((2,2), (0,0))
plt.scatter(x/3,y/3,color='blue',s = 10)
plt.scatter([0],[0],label='Close', color='red')
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xticks([])
plt.yticks([])

plt.subplot2grid((2,2), (0,1))
plt.scatter(x/3+2,y/3+2,color='blue',s = 10)
plt.scatter([0],[0],label='Close', color='red')
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xticks([])
plt.yticks([])

plt.subplot2grid((2,2), (1,0))
plt.scatter(x,y,color='blue',s = 10)
plt.scatter([0],[0],label='Close', color='red')
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xticks([])
plt.yticks([])

plt.subplot2grid((2,2), (1,1))
plt.scatter(x+2,y+2,color='blue',s = 10)
plt.scatter([0],[0],label='Close', color='red')
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xticks([])
plt.yticks([])

plt.show()
```

Each sub-plot represents a different model, some models are good, some are bad. But they all make errors in a different way. Here we break down the types of errors a model can make into two terms:

1. Bias: The inability of the model to learn a representation which fits the data. A high bias means a model is not powerful enough to find the patterns in the data, it learns poor representations.
2. Variance: The sensitivity of the model to changes in the data. A high variance model means it is too powerful to fit the patterns in the data and has instead fitted to the noise. A model with high variance has massive swings in its predictions with only minor changes in the underlying data.

Some of these models have low bias and low variance some have high bias and high variance, some have a mixture of the two. How would you label the error of each model here?

### Bias  Variance Tradeoff with KNN

When optimizing our models we often encounter a tradeoff in the error from bias and variance. We can decrease the error due to one but often at a cost of more error from the other. This tradeoff is formally known as the Bias Variance Tradeoff. 

Let's actually explore and visualize the error Bias and Varaince with a KNN model, we'll see how the tradeoff in error changes with our model parameters.

We will fit the model to the following data:

```python
x = np.arange(-1,1,.1)
y = np.sqrt(1-(x**2))

data = np.transpose([np.tile(x, len(x)), np.repeat(x, len(x))])

label = np.where(data[:,1]+1/2 >= np.sqrt(np.abs(1/200-data[:,0]**2)),1,0)
plt.scatter(data[:,0],data[:,1],c=label);
```

The ideal model boundary is pretty clear, the boomerang shape in the middle, but what if we didn't know this? Also what if our data was a bit noisy:

```python
x = np.arange(-1,1,.1)
y = np.sqrt(1-(x**2))

np.random.seed(100)

data = np.transpose([np.tile(x, len(x)), np.repeat(x, len(x))])

label = np.where(data[:,1]+1/2 >= np.sqrt(np.abs(1/200-data[:,0]**2)),1,0)
index = np.random.choice(len(label),75,replace=False)
label[index] = 1-label[index]

plt.scatter(data[:,0],data[:,1],c=label)

plt.show()
```

Let's fit a KNN model where $K=1$:

```python
from sklearn.neighbors import KNeighborsClassifier

oneNN_one = KNeighborsClassifier(n_neighbors = 1).fit(data,label)
```

Let's also fit another KNN model with $K=1$ to another set of noisy data:

```python
x = np.arange(-1,1,.1)
y = np.sqrt(1-(x**2))

np.random.seed(12)

data = np.transpose([np.tile(x, len(x)), np.repeat(x, len(x))])

label = np.where(data[:,1]+1/2 >= np.sqrt(np.abs(1/200-data[:,0]**2)),1,0)
index = np.random.choice(len(label),75,replace=False)
label[index] = 1-label[index]

plt.scatter(data[:,0],data[:,1],c=label)

plt.show()

oneNN_two = KNeighborsClassifier(n_neighbors = 1).fit(data,label);
```

```python
PlotBoundaries(oneNN_one,data)
PlotBoundaries(oneNN_two,data)
```

Despite the fact the models are fitted to near identical data we see their predictions are quite different. With $K=1$ we end up with a high variance model.

On the flip side if we fit a KNN model with a really high K value to two samples of noisy data and plot the boundaries:

```python
x = np.arange(-1,1,.1)
y = np.sqrt(1-(x**2))

np.random.seed(100)

data = np.transpose([np.tile(x, len(x)), np.repeat(x, len(x))])

label = np.where(data[:,1]+1/2 >= np.sqrt(np.abs(1/200-data[:,0]**2)),1,0)
index = np.random.choice(len(label),75,replace=False)
label[index] = 1-label[index]

big_KNN_one = KNeighborsClassifier(n_neighbors = len(label)-50).fit(data,label);

x = np.arange(-1,1,.1)
y = np.sqrt(1-(x**2))

np.random.seed(12)

data = np.transpose([np.tile(x, len(x)), np.repeat(x, len(x))])

label = np.where(data[:,1]+1/2 >= np.sqrt(np.abs(1/200-data[:,0]**2)),1,0)
index = np.random.choice(len(label),75,replace=False)
label[index] = 1-label[index]

big_KNN_two = KNeighborsClassifier(n_neighbors = len(label)-50).fit(data,label);
```

```python
PlotBoundaries(big_KNN_one,data)
PlotBoundaries(big_KNN_two,data)
```

We see the predictions here are more similar than before, the variance is down quite a bit. The model is still not good, it overestimates the purple class quite a bit, it has a high bias.

We can also compare the predictions of the models to see the breakdown because of variance and bias:

```python
print("High bias models agree: %3.3f percent of the time" % (100*compare_models(big_KNN_one,big_KNN_two)))
```

```python
print("High variance models agree: %3.3f percent of the time" % (100*compare_models(oneNN_one,oneNN_two)))
```

As expected the high bias, but low variance, models agree most of the time. While the high variance, but low bias, models agree much less frequently.

At this point you may be wondering if any of the models are better than the others, if we test out how they do on the "ideal" data set:

```python
x = np.arange(-1,1,.1)
y = np.sqrt(1-(x**2))

data = np.transpose([np.tile(x, len(x)), np.repeat(x, len(x))])

label = np.where(data[:,1]+1/2 >= np.sqrt(np.abs(1/200-data[:,0]**2)),1,0)
index = np.random.choice(len(label),75,replace=False)
label[index] = 1-label[index]

print(big_KNN_one.score(data,label))
print(big_KNN_two.score(data,label))
print(oneNN_one.score(data,label))
print(oneNN_two.score(data,label))
```

We see that they are all almost as good, or bad, as each other.

So what does this mean from a performance perspective? Since our total error is affected by both terms optimizing to minimize bias or variance alone is not a good idea. Instead we must trade one for the other and end up at some ideal middle ground. It can help to visualize the error as a plot:

```python
complexity = np.arange(2,12,.001)
bias = 1 / (complexity**1.01)
variance = (complexity)**2 / 200

total_error = bias+variance

min_total = total_error.argmin()

plt.plot(complexity,bias, label="bias")
plt.plot(complexity,variance, label="variance")
plt.plot(complexity,total_error, label="total error")

plt.scatter(complexity[min_total],total_error[min_total], color = "gold")

plt.legend()
plt.xlabel("Complexity")
plt.ylabel("Error")

plt.show()
```

Essentially we are trying to tune our model complexity so that we end up at the gold dot, this is the best model we can hope for. In reality we don't know how the error due to bias and variance change as we adjust the model complexity, we know that they decrease and increase as the models become more complex but we don't know the exact pattern to the change.

This has been a bit of a theoretical section on where errors come from, but it should help you understand that a model that is too complex (high variance low bias) is not good, while a model that is too simple (high bias low variance) is also not good. In the language of machine learning you can think of this as finding a good inbetween of a model that overfits (high variance) and a model that underfits (high bias).

### ROC and AUC

The final model evaluation metric we will use is the *Receiver Operating Characteristic*, or ROC. The ROC is a graphical measure which compares the true positive and false positive rate of a binary model.

Let's start with a simple example. Recall, logistic regression makes its decision about the class of some data point $x$ by computing the sigmoid function:

$$\sigma(x) = \frac{1}{1+e^{-\beta \cdot x}}$$

And comparing this value to some threshold, we usually set this threshold at $0.5$. More generally for any model this $0.5$ threshold means we put a point in class $1$ when our model beleives it is slightly more likely to belong to class $1$. But what happens if we adjust this threshold? 

If we set the threshold lower, say $0.4$, we will get more class $1$ examples correct but we will also now incorrectly classify some class $0$ examples as class $1$. That is we increase the true positive rate and false positive rate. What if we do the reverse, and set the threshold at 0.6? We would get the reverse, both the true and false positive rates would decrease. 

The ROC is a direct comparison between the true postive and false positive rates at various threshold levels. Let's actually look at one such curve (the code for this example is a bit complex, so we'll work through it slowly):

```python
from sklearn.metrics import roc_curve, auc
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer

# Use the cancer dataset
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target

# Add some noise to the data
random_state = np.random.RandomState(10)
n_samples, n_features = X.shape
X = X + (random_state.randn(n_samples, n_features) * 1000)

# Create a train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3,
                                                    random_state=0)

# Train one vs rest SVMs. Need to use the OneVsRestClassifier for ROC/AUC.
classifier = OneVsRestClassifier(LogisticRegression(random_state=100))

#Get the probability for each point in the test set.
probs = classifier.fit(X_train, y_train).predict_proba(X_test)

# Compute ROC curve and AUC for for the one class
fpr, tpr, _ = roc_curve(y_test, probs[:, 1])
roc_auc = auc(fpr, tpr)
  
# Plot the ROC curve.
plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='AUC = %0.2f' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC and AUC')
plt.legend(loc="best")
plt.show()
print("AUC score: %3.3f" % roc_auc)
```

Each point on the curve represents a tradeoff between true and false positive rates. If we're only willing to tolerate a false positive rate of $0.2$ then we can only achieve a true positive rate of $0.6$, on the other hand if we're willing to tolerate a false positive rate of $0.8$ we can correctly classify all positive examples (a true positive rate of $1.0$). The diagonal line represents a random guess. It serves as a baseline to compare the curve against. When the curve is above it we are doing better than random, below we are doing worse.

Now that we have this curve what can we do with it? It is usefull for tunning acceptance thresholds. Say our test was measuring if an aeroplane required maitnance. We'd much rather have a perfect true positive rate and several false positives than miss classify aeroplanes that were going to fail.

It can also provide a measure of model quality. We can do this by measuring the area under the ROC curve (or AUC score). 

To understand why this measure is usefull consider the following three examples. The best model will never make a missclassification, so it will have a true positive rate of $1.0$ and a false positive rate of $0$. It's ROC curve is the flat horizontal line at 1. The area under it is $1$. The worst classifier is then the horrizontal line at 0, it has the inverse true and false positive rate of the best model, its AUC score is $0$. The random guessing classifier has an AUC score of $0.5$. The key observation here is a higher AUC score corresponds to a better ROC curve, thus we can measure the quality of a model with its AUC score.

Two important notes. 
1. ROC and AUC values are data dependent. If we change the underlying distribution, or even the points from the same distribution, the plots and values can significantly change.
2. We found the ROC curve and AUC score by fitting a model to a training set and evaluating it on a validation set. If we had instead fitted and evaluated on the same set then we would likely be  making our evaluation and decisions on data we have overfitted to. This is akin to selecting the best performing model by using the accuracy on the training set.

```python
#Get the probability for each point in the train set.
probs = classifier.predict_proba(X_train)

# Compute ROC curve and AUC for for the one class
fpr, tpr, _ = roc_curve(y_train, probs[:, 1])
roc_auc = auc(fpr, tpr)
  
# Plot the ROC curve.
plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='AUC = %0.2f' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC and AUC')
plt.legend(loc="best")
plt.show()
print("AUC score: %3.3f" % roc_auc)
```

As expected the AUC score is higher.

### Using ROC and AUC

Besides using AUC (and hence the ROC) as an evaluation criteria is there anything else we can do with it? While AUC is usually correlated with accuracy it is possible that we can choose between a model with a high AUC and lower accuracy and a model with a lower AUC and high accuracy. Let's step back and see what each is measuring again:

1. Accuracy is measuring the fraction of points our model gets correct for a fixed threshold.
2. AUC is the probability that when our model is given two random points, one from the positive class and one from the negative class, our model is more confident that the positive point belongs to the positive class than the negative point belonging to the positive class.

We've seen accuracy is not always a great measure when our classes or highly skewed, this can lead us to a false sense of confidence. With AUC we are protected from class imbalances, consider the hospital dataset from the start of machine learning, we saw linear models could get high accuracy on it but the predictions were not quite what we expected, if we plot out the ROC and AUC on the data set we get:

```python
x_one = np.random.randn(50)*3 + 10
x_two = np.random.randn(20)*3 + 60
x_three = np.random.randn(50)*3 + 200
X = np.concatenate((x_one,x_two,x_three))

X = X.reshape(-1,1)

y_one = np.full((50),1)
y_two = np.full((20),0)
y_three = np.full((50),1)
                  
y = np.concatenate((y_one,y_two,y_three), axis=0)
```

```python
classifier = OneVsRestClassifier(LogisticRegression(random_state=100))

#Get the probability for each point in the train set.
probs = classifier.fit(X,y).predict_proba(X)

# Compute ROC curve and AUC for for the one class
fpr, tpr, _ = roc_curve(y, probs[:, 1])
roc_auc = auc(fpr, tpr)
  
# Plot the ROC curve.
plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='AUC = %0.2f' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC and AUC')
plt.legend(loc="best")
plt.show()
print("AUC score: %3.3f" % roc_auc)
```

First notice the AUC score is saying our model is no better than random guessing.

Also, if we look at the ROC curve we can conclude if we aren't willing to tolerate a false positive rate of 100% then we will only get a 50% true positive rate.

ROC and AUC are just another tool for evaluating our models, it is just one more piece of information when doing evaluation. That being said compared to pure accuracy scores they do provide more details into the inner workings of our models.
