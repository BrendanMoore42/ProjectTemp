# Unsupervised Learning - KMeans and Mixture of Gaussians

All of the statistics and machine learning modeling we've been doing has been on data with an independent variable (a set of observations or features) and dependent varaiables (an outcome). Our goal can usually be phrased as there is some effect which causes the independent variable to give rise to the dependent variable, how do we model and understand this effect? Up until now we've assumed we know what the dependent variable looks like, we know the range of values it can take and what values it takes for the data we fit our model to (the training set), this is broadly known as supervised learning. Today we will see unsupervised learning, we will learn to fit models where we do not the value of the dependent variable, we won't even assume we know the range it comes from.

As always let's import some common packages:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk

from sklearn.metrics import accuracy_score
from sklearn import model_selection, metrics, linear_model, datasets, feature_selection
```

## Unlabeled Data

Unlabeled data, data where we do not know the values of the dependent variable, can be much more frequent that expected. There are a few resons for why data can be unlabeled:

1. The data was improperly gathered (missing or incorrect labels).
2. There is no label on the data but there are still possible patterns that can be found.
3. There is an underlying pattern in the data, but we've never encountered it, and thus don't know how to annotate it.

Whatever the reasons for it, we need to find ways to deal with and analyze unlabeled data.

### Clustering

Clustering involves fitting a model which puts similar data into the same group. Classification (fitting a model with a dependent catagorical variable) can be thought of as a form of clustering, we want our model to put all points with the same dependent variable label into the same group.

Besides categorical labels there can be other measures of similarity, we can group people based on height or customers on their spending habbits. As we will see in the next section we can have several natural notions of what reasonable clusters are.

### Clustering  Unlabeled Data:

Now we will examine how to cluster unlabeled data. Essentially this means we want to partition the unlabeled data into similar groups (if the data were labeled then the groups should be obvious).

First let's generate some unlabeled data (for ease of visualizaiton we'll restrict ourselves to two dimensions):

```python
points_per_class = 100

#Generate four clusters.
mean = [5,5]
cov = [[1,0],[0,1]]
x_one = np.random.multivariate_normal(mean, cov, (points_per_class))
y_one = np.full((points_per_class),0)

mean = [-5,-5]
cov = [[1,0],[0,1]]
x_two = np.random.multivariate_normal(mean, cov, (points_per_class))
y_two = np.full((points_per_class),0)

mean = [0,0]
cov = [[0.25,0],[0,0.25]]
x_three = np.random.multivariate_normal(mean, cov, (points_per_class))
y_three = np.full((points_per_class),1)

mean = [4,-2.5]
cov = [[0.25,0],[0,0.25]]
x_three = np.random.multivariate_normal(mean, cov, (points_per_class))
y_three = np.full((points_per_class),0)

mean = [-6,2.5]
cov = [[0.25,0],[0,0.25]]
x_four = np.random.multivariate_normal(mean, cov, (points_per_class))
y_four = np.full((points_per_class),0)

X = np.concatenate((x_one,x_two,x_three,x_four),axis=0)
Y = np.concatenate((y_one,y_two,y_three,y_four),axis=0)

#Plot the data
plt.scatter(X[:, 0], X[:, 1], c=Y, s=20, edgecolor='k')

plt.show()
```

Even though we don't have class labels (colors) for each data point there is an intuitive split the data, it appears there are four distinct clusters, and it seems reasonable each would belong to a different class of data.

```python
#Add labels to the data.
y_one = np.full((points_per_class),1)
y_two = np.full((points_per_class),2)
y_three = np.full((points_per_class),3)
y_four = np.full((points_per_class),4)

Y = np.concatenate((y_one,y_two,y_three,y_four),axis=0)

#Plot
plt.scatter(X[:, 0], X[:, 1], c=Y, s=20, edgecolor='k')
plt.show()
```

That example was easy to reason about, we could visualize the data and the clusters had pretty clear boundaries. But what happens if the distinction isn't so obvious? What about this data, how many clusters are there?

```python
#Generate some data
mean = [3,3]
cov = [[1,0],[0,1]]
x_one = np.random.multivariate_normal(mean, cov, (points_per_class))
y_one = np.full((points_per_class),0)

mean = [1,1]
cov = [[1,0],[0,1]]
x_two = np.random.multivariate_normal(mean, cov, (points_per_class))
y_two = np.full((points_per_class),0)

mean = [-3,-3]
cov = [[1,0],[0,1]]
x_three = np.random.multivariate_normal(mean, cov, (points_per_class))
y_three = np.full((points_per_class),0)

mean = [-1,-1]
cov = [[1,0],[0,1]]
x_four = np.random.multivariate_normal(mean, cov, (points_per_class))
y_four = np.full((points_per_class),0)

X = np.concatenate((x_one,x_two,x_three,x_four),axis=0)
Y = np.concatenate((y_one,y_two,y_three,y_four),axis=0)

#Plot
plt.scatter(X[:, 0], X[:, 1], c=Y, s=20, edgecolor='k')
plt.show()
```

How many classes are there, and where are they? This isn't as obvious there are a few equally good explanations.

```python
#One possible answer.
y_one = np.full((points_per_class),1)
y_two = np.full((points_per_class),2)
y_three = np.full((points_per_class),3)
y_four = np.full((points_per_class),4)
Y = np.concatenate((y_one,y_two,y_three,y_four),axis=0)

plt.scatter(X[:, 0], X[:, 1], c=Y, s=20, edgecolor='k')
plt.show()

#Another possible answer.
y_one = np.full((points_per_class),1)
y_two = np.full((points_per_class),1)
y_three = np.full((points_per_class),2)
y_four = np.full((points_per_class),2)
Y = np.concatenate((y_one,y_two,y_three,y_four),axis=0)

plt.scatter(X[:, 0], X[:, 1], c=Y, s=20, edgecolor='k')
plt.show()

#Another^2 possible answer.
y_one = np.full((points_per_class),1)
y_two = np.full((points_per_class),3)
y_three = np.full((points_per_class),2)
y_four = np.full((points_per_class),2)
Y = np.concatenate((y_one,y_two,y_three,y_four),axis=0)

plt.scatter(X[:, 0], X[:, 1], c=Y, s=20, edgecolor='k')
plt.show()
```

These all seem like reasonable answers for the number of classes and cluster locations. How do we know which is correct? Can we even answer this question? Furthermore, how dow we find any of these clusters?

## The KMeans algorithm

The first algorithm we will see for unsupervised clustering is the *KMeans* algorithm. The algorithm can be found under `sklearn.cluster` (for full documentation see: http://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html). The only parameter we need to focus on is the number of clusters `n_clusters` (it has a default value of 8). Running KMeans is no differnt than running any other model in Scikit-Learn, we build the model, fit it to data then make prediction on unseen data. The prediction of our model on a data point is the cluster it beleives the point belongs to.

First let's take the previous example without any labels:

```python
points_per_class = 100

#Generate four clusters.
mean = [5,5]
cov = [[1,0],[0,1]]
x_one = np.random.multivariate_normal(mean, cov, (points_per_class))
y_one = np.full((points_per_class),0)

mean = [-5,-5]
cov = [[1,0],[0,1]]
x_two = np.random.multivariate_normal(mean, cov, (points_per_class))
y_two = np.full((points_per_class),0)

mean = [0,0]
cov = [[0.25,0],[0,0.25]]
x_three = np.random.multivariate_normal(mean, cov, (points_per_class))
y_three = np.full((points_per_class),1)

mean = [4,-2.5]
cov = [[0.25,0],[0,0.25]]
x_three = np.random.multivariate_normal(mean, cov, (points_per_class))
y_three = np.full((points_per_class),0)

mean = [-6,2.5]
cov = [[0.25,0],[0,0.25]]
x_four = np.random.multivariate_normal(mean, cov, (points_per_class))
y_four = np.full((points_per_class),0)

X = np.concatenate((x_one,x_two,x_three,x_four),axis=0)
Y = np.concatenate((y_one,y_two,y_three,y_four),axis=0)

#Plot the data
plt.scatter(X[:, 0], X[:, 1], c=Y, s=20, edgecolor='k')

plt.show()
```

Next let's fit the KMeans model, in this case we'll specify four clusters (since this seems like a reasonable number for this data):

```python
from sklearn.cluster import KMeans

#Specify the model
k_means_model = KMeans(n_clusters = 4)
#Fit the model to the data
k_means_model.fit(X);
```

Now we can make predictions about each group of data points:

```python
#Annotate the data points with the KMeans prediction
Y = k_means_model.predict(X)
#Plot the data:
plt.scatter(X[:, 0], X[:, 1], c=Y, s=20, edgecolor='k')
plt.show()
```

This fit is pretty good, it found the natural (at least natural to us) clusters. What if we ran it on the more ambiguous example?

### What is KMeans doing

It seems on the previous examples KMeans performed quite well, but do we have an objective measure of what is good? To answer this we need to understand what KMeans is doing. The model is fitting $K$ (`n_clusters`) special points to the data, each of these special points represents a distinct label. These special points are called the centers, and each is placed in the center of some cluster of data. The algorithm to find the best centers is actually quite complex, see: https://en.wikipedia.org/wiki/Expectation%E2%80%93maximization_algorithm for the idea behind it. When making a prediction for the label of a datapoint $x$  we assign it to the same label as the nearest centre.

If we re-ran our KMeans model on the previous (simple) data, but this time plotted the centers.

```python
points_per_class = 100

mean = [5,5]
cov = [[1,0],[0,1]]
x_one = np.random.multivariate_normal(mean, cov, (points_per_class))

mean = [-5,-5]
cov = [[1,0],[0,1]]
x_two = np.random.multivariate_normal(mean, cov, (points_per_class))

mean = [4,-2.5]
cov = [[0.25,0],[0,0.25]]
x_three = np.random.multivariate_normal(mean, cov, (points_per_class))

mean = [-6,2.5]
cov = [[0.25,0],[0,0.25]]
x_four = np.random.multivariate_normal(mean, cov, (points_per_class))

X = np.concatenate((x_one,x_two,x_three,x_four),axis=0)

#Specify the model
k_means_model = KMeans(n_clusters = 4)
k_means_model.fit(X)


#Annotate the data points with the KMeans prediction
Y = k_means_model.predict(X)

#Plot
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolor='k')

#Plot the K centers
plt.scatter(k_means_model.cluster_centers_[:, 0],k_means_model.cluster_centers_[:, 1], c='red',marker="*",s=200)
plt.show()
```

We can see each center (the red stars) gets assigned to the middle of a data cluster. For illustrative purpouses we can adjust two of the centers so that the data label assignment changes:

```python
#Change the center locations (finds a bad fit)
k_means_model.cluster_centers_ = np.array([[ 4.93629851,  4.86487333],\
                                           [-0, -6],\
                                           [ 3.98048973, -2.49005718],\
                                           [-0,  -4]])

#Annotate the data points with the KMeans prediction
Y = k_means_model.predict(X)

#Plot
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolor='k')

#Plot the K centers
plt.scatter(k_means_model.cluster_centers_[:, 0],k_means_model.cluster_centers_[:, 1], c='red',marker="*",s=200)
plt.show()
```

Now some of the data in the bottom left cluster is assigned to a new cluster (the same as the cluster directly above it), since the top left center has become closer.

### How to evaluate the fit

The last fit we saw seems poorer that the one before and from the perspective of the KMeans objective function it is. When fitting a KMeans model we aim to minimize the sum of square distance between each point and its assigned cluster center. Formally if we have $d$ features and $x$ is a datapoint and $c$ is the nearest center, we assign $x$ to class $c$ and the distance between the two is:

$$dist(x,c) = \sum_{i=1}^d (x_i - y_i)^2$$

The total score of our fit is then just the sum of: 

$$dist(x',c')$$

For every data point $x'$ and its assigned (or nearest) cluster center $c'$. The goal is to minimize this distance. Actually finding the centers which minimize the distance is very hard (and it is suspected there is no fast way to do so). Instead there are several heuristic approaches which seem to do well in practice. The actual distance score of our fit is stored in the `inertia_` variable, you can get the inertia value as follows:

```python
print('The fit score (inertia) is: %3.3f' % k_means_model.inertia_)
```

### Data Scale

Just like with KNN, KMeans is a distance based model, because of this the scale of data can have a huge impact. Recall the distance between two points is:

$$dist(x,y) =  \sqrt{\sum_{i=1}^d(x_i - y_i)^2}$$

If our features lie in widely different scales then the impact of some features will be washed away. Because of this it is not uncommon to scale the data prior to passing it into KMeans.

### How to Pick K

In this section we will discuss how to find the "ideal" number of clusters. What the appropriate number is can be highly subjective. Let's look at a few options:

1. There is a categorical dependent variable and we know how many underlying classes there are. In this case we can set the K appropriately. If we had the cancer dataset we know ther are two classes, hence we can fit 2 clusters.
2. We want to segment the data into a predetermined number of classes. If we had retail data and suspected, or wanted to divide, our customers into a predetermined number of types we could use this number for our K.
3. We can use the inertia value to help guide our selection (but how?)

Let's load up the following three cluster data:

```python
points_per_class = 20

mean = [5,5]
cov = [[1,0],[0,1]]
x_one = np.random.multivariate_normal(mean, cov, (points_per_class))

mean = [-5,-5]
cov = [[1,0],[0,1]]
x_two = np.random.multivariate_normal(mean, cov, (points_per_class))

mean = [4,-2.5]
cov = [[0.25,0],[0,0.25]]
x_three = np.random.multivariate_normal(mean, cov, (points_per_class))

X = np.concatenate((x_one,x_two,x_three,x_four),axis=0)
```

And plot out the inertia values (This will take a bit of time...)

```python
k_range = np.arange(1,20)

inertia_list = []

for k in k_range :
    
    #Specify the model
    k_means_model = KMeans(n_clusters = k)
    k_means_model.fit(X)
    
    inertia_list.append(k_means_model.inertia_)
    
plt.scatter(k_range,inertia_list)
plt.show()
```

This looks like it took a pretty hard dip right away and levelled off. If we focus on the first few points:

```python
plt.scatter(np.arange(7),inertia_list[0:7])
```

If you notice after the third/fourth dot the values level off quite a bit, this mean seach new centroid is not reducing the inertia by much. This leads evidence to the idea the ideal number of classes is three or four. 

This isn't very scientifc, but it can be a nice approximation. How we draw the cutoff is hard to determine, it relies on some judgement and further investigation.

### Unseen Data and What's the Point (of KMeans)

So far we've only visualized what KMeans predicted for data it was fitted on. What happens when we give it unseen data? While this may appear similar to a test set it's not quite the same, we're not trying to score how our model generalizes to new data, instead we're trying to see which of our discovered clusters some new data belongs to.

Let's load up the Iris data (and ignore the label for now):

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=1)
```

First let's find out how many clusters we should try out:

```python
k_range = np.arange(1,10+1)

inertia_list = []

for k in k_range :
    
    #Specify the model
    k_means_model = KMeans(n_clusters = k)
    k_means_model.fit(X_train)
    
    inertia_list.append(k_means_model.inertia_)
    
plt.scatter(k_range,inertia_list)
plt.show()
```

It looks like it's levelled off pretty quickly, maybe 2 or 3 clusters, let's try out a model for each idea:

```python
two_means_model = KMeans(n_clusters = 2)
two_means_model.fit(X_train);

three_means_model = KMeans(n_clusters = 3)
three_means_model.fit(X_train);
```

Let's visualize some of the learned means (we'll do this by plotting pairs of features):

```python
y_predict = two_means_model.predict(X_train)
plt.scatter(X_train[:,0],X_train[:,2],c=y_predict)
plt.title("2 Means")
plt.show()

y_predict = three_means_model.predict(X_train)
plt.scatter(X_train[:,0],X_train[:,2],c=y_predict)
plt.title("3 Means")
plt.show()
```

Both of these clusterings seem plausable, if you look at the actual Iris labels the three group clustering is quite good.

Now if we encounter more data, say we came across a new flower and wamted to classify it, we can just use our model to assign the new point to one of the discovered clusters:

```python
y_predict = two_means_model.predict(X_test)
plt.scatter(X_test[:,0],X_test[:,2],c=y_predict)
plt.title("2 Means")
plt.show()

y_predict = three_means_model.predict(X_test)
plt.scatter(X_test[:,0],X_test[:,2],c=y_predict)
plt.title("3 Means")
plt.show()
```

Why do we want to cluster these new data points / why do we want to cluster the old data points? The ability to put our data in groups with like data points can be very usefull.

1. If we lacked data labels this can serves as a good alternative for classification.
2. Once we assign data to "natural" cluster we can use these clusters to help guide our models. For example if we were clustering shoppers we may find one cluster represents high volumne shoppers while another contains new parents. This allows us to find patterns for marketing.
3. We can use clustering to help generate features for other algorithms (this is a pretty advanced topic).

## Mixture of Gaussians

One of the major limiting factors to KMeans is each feature dimension equally contributes to the score and thus has an equal impact on the fitting of the model. This leads to circular groupings (or spherical in higher dimensions). 

But our data classes are not always circular in nature, consider the following two classes generated with multi variable gaussians. While we won't go into what this means exactly (see: https://en.wikipedia.org/wiki/Multivariate_normal_distribution) the important point is one of the classes is more of an oval than a circle.

```python
#Generate some data
points_per_class = 300

mean = [2,2]
cov = [[1,0],[0,1]]
x_one = np.random.multivariate_normal(mean, cov, (points_per_class))

mean = [-2,-2]
cov = [[1,10],[1,1]]
x_two = np.random.multivariate_normal(mean, cov, (points_per_class))

X = np.concatenate((x_one,x_two),axis=0)

y_one = np.full(points_per_class,1)
y_two = np.full(points_per_class,0)
Y = np.concatenate((y_one,y_two),axis=0)

#Plot
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolor='k')
plt.show()
```

If we try and fit a KMeans classifier with two centers we get:

```python
#Specify the model
k_means_model = KMeans(n_clusters = 2)
k_means_model.fit(X)

#Annotate the data points with the KMeans prediction
y_one = k_means_model.predict(x_one)
y_two = k_means_model.predict(x_two)
Y = np.concatenate((y_one,y_two),axis=0)

plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolor='k')
plt.scatter(k_means_model.cluster_centers_[:, 0],k_means_model.cluster_centers_[:, 1], c='red',marker="*",s=200)

plt.show()
```

While the fit isn't implausable it doesen't match the underlying distributions we generated the data with. 

Instead we can fit a *Gaussian Mixture Model* (GMM). A GMM is essentially the same as a KMeans model, but it allows for non circular classes. To specify a GMM we use the `GaussianMixture` from `sklearn.mixture`. Like KMeans we need to specify the number of classes we are fitting (`n_components`) and set `covariance_type='full'` (this allows it to fit non circular shapes). Fitting and making predictions on the data is exactly the same as before.

```python
from sklearn.mixture import GaussianMixture

#Build and fit the model
my_gmm = GaussianMixture(n_components=2, covariance_type='full')
my_gmm.fit(X)

#Make predictions
y_one = my_gmm.predict(x_one)
y_two = my_gmm.predict(x_two)
Y = np.concatenate((y_one,y_two),axis=0)

#Plot the predictions (unlike with KMeans the GMM in sklearn does not include a way to get the middle)
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolor='k')
plt.show()
```

This is pretty good, it has learned to fit the data in a way that matches the underlying distribution which generated it.

The considerations we had with KMeans (when to apply it/what K to select) are essentially the same here. If we want to apply a GMM we need to appply our domain knowledge to help guide our decisions. Also it is worth noting KMeans is often sufficient in practice. While GMMs can fit more complex data it is usually the case KMeans gets the job done to a satisfactory level.
